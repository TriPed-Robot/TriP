```python
import matplotlib.pyplot as plt
import csv
```

```python
def read_csv(filename, dtype=float):
    with open(filename) as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        first = True
        lst = []
        rows = 0
        for row in reader:
            if first:
                rows = len(row)
                for i in range(rows):
                    lst.append([])
                first = False

            for j in range(rows):
                lst[j].append(float(row[j]))
        return lst

def get_error(data_1,data_2):
    error = []
    for i in range(len(data_1)):     
        error.append(abs(data_2[i] - data_1[i]))
    return error
```

```python
foot_coordinates = read_csv('output_data/forward_foot_coordinates.csv')
matlab_foot_coordinates = read_csv('matlab_output/matlab_foot_coordinates.csv')
#tips = read_csv('matlab_output/gimbal_extend.csv')
```

```python
foot_coordinates_x = foot_coordinates[0]
foot_coordinates_y = foot_coordinates[1]
foot_coordinates_z = foot_coordinates[2]

matlab_foot_coordinates_x = matlab_foot_coordinates[1]
matlab_foot_coordinates_y = matlab_foot_coordinates[2]
matlab_foot_coordinates_z = matlab_foot_coordinates[3]

#tip_t1 = tips[1]
#tip_t2 = tips[2]
#tip_e = tips[3]

#time = matlab_foot_coordinates[0]

error_t1 = get_error(matlab_foot_coordinates_x,foot_coordinates_z)
error_t2 = get_error(matlab_foot_coordinates_y,foot_coordinates_y)
error_e = get_error(matlab_foot_coordinates_z,foot_coordinates_z)
```

```python
plt.plot(foot_coordinates_x,label="my_x")
plt.plot(foot_coordinates_y,label="my_y")
plt.plot(foot_coordinates_z,label="my_z")
plt.plot(matlab_foot_coordinates_x,label="matlab_x")
plt.plot(matlab_foot_coordinates_y,label="matlab_y")
plt.plot(matlab_foot_coordinates_z,label="matlab_z")


plt.title("Exp_1_foot_coordinates")
plt.legend(loc='upper right')
#plt.ylim(-1.5,1.5)
plt.show()
```

```python
plt.plot(error_t1,label="error_t1")
plt.plot(error_t2,label="error_t2")
plt.plot(error_e,label="error_e")
plt.title("Exp_1 Error")
plt.legend(loc='upper right')
plt.show()
```
